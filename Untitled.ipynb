```python
def extract_cpu_brand(text):
    text=text.lower()
    if 'intel core i3' in text:
        return 'Intel core i3'
    elif 'intel core i5' in text:
        return 'Intel core i5'
    elif 'intel core i7' in text:
        return 'Intel core i7'
    elif 'intel' in text:
        return 'Other intel processor'
    elif 'amd' in text:
        return 'AMD processor'
    else:
        return 'unknown'
df['Cpu brand']=df['Cpu'].apply(extract_cpu_brand)       
```

```python
import re
def extract_val(parts):
    parts=parts.strip()
    parts=parts.upper()
    partss=re.match(r"(\d+\.?\d*)TB",parts)
    if partss:
        return float(partss.group(1))*1024
    partss=re.match(r"(\d+\.?\d*)GB",parts)
    if partss:
        return float(partss.group(1))
    return 0
def passe(strr):
    strr=strr.lower()
    parts=strr.split('+')
    SSD=HDD=Flash_Storage=Hybrid=0
    for part in parts:
        if 'ssd' in part:
           SSD+=extract_val(part)
        if 'hdd' in part:
           HDD+=extract_val(part)
        if 'hybrid' in part:
           Hybrid+=extract_val(part)
        if 'flash storage' in part:
           Flash_Storage+=extract_val(part)
    return pd.Series([SSD, HDD, Flash_Storage, Hybrid])    
        
```

```python
df[['SSD', 'HDD', 'Flash_Storage', 'Hybrid']] = df['Memory'].apply(passe) 
```

```python
x=df.drop(columns=['Price'])
y=np.log(df['Price'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train
```

```python
step1=ColumnTransformer(transformers=[('colm_t',OneHotEncoder(drop='first',sparse_output=False),[0,1,7,10,11])],remainder='passthrough')
```

```python
#DecisionTreeRegressor
step2=DecisionTreeRegressor(max_depth=8)
pipe=Pipeline([('step1',step1),
               ('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print('R2_score',r2_score(y_test,y_pred))
```

```python
#RandomForestRegressor
step2=RandomForestRegressor(n_estimators=100,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)
pipe=Pipeline([('step1',step1),
               ('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print('R2_score',r2_score(y_test,y_pred))
```
